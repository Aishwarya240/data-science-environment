## Dot product example(nut butter sales)

```python
np.random.seed(0)
# Number of jars sold
sales_amounts = np.random.randint(20, size=(5,3))
sales_amounts
```

```python
# Create weekly_sales Dataframes
weekly_sales = pd.DataFrame(sales_amounts,
                            index=["Mon", "Tues", "Wed", "Thurs", "Fri"],
                            columns=["Almond butter", "Peanut butter", "Cashew Butter"])
weekly_sales
```

```python
# Create prices array
prices = np.array([10, 8, 12])
prices
```

```python
 prices.shape
```

```python
sales_amounts.shape
```

```python
# Create butter_prices Dtaframe
butter_prices = pd.DataFrame(prices.reshape(1, 3),
                             index=["Price"],
                             columns=["Almond butter", "Peanut butter", "Cashew butter"])

butter_prices
```

```python
total_sales = prices.dot(sales_amounts)
```

```python
#Shapes arent aligned, lets transpose
total_sales = prices.dot(sales_amounts.T)
total_sales
```

```python
#Create daily_sales
butter_prices
```

```python
weekly_sales
```

```python
butter_prices.shape, weekly_sales.shape
```

```python
weekly_sales.T.shape
```

```python
daily_sales = butter_prices.values.dot(weekly_sales.values.T)
daily_sales.shape
```

```python
weekly_sales.shape
```

```python
# Option 1: print explicitly
print(daily_sales)

# Option 2: convert to DataFrame for labels
daily_sales_df = pd.DataFrame(daily_sales,
                              index=butter_prices.index,
                              columns=weekly_sales.index)  # assuming weekly_sales index = days
daily_sales_df
```

```python
weekly_sales
```

```python
#Doesnt work, not the right shape
weekly_sales["Total ($)"] = daily_sales
weekly_sales
```

```python
weekly_sales["Total ($)"] = daily_sales.T
weekly_sales
```
