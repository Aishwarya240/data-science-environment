```python
import pandas as pd

# Load CSV
car_sales = pd.read_csv("car_sales.csv")

# Check first 5 rows
print(car_sales.head())
```

```python
# Check first few rows
print(car_sales.head())
```

```python
# Get column names, data types, and non-null counts
print(car_sales.info())

# Get summary statistics for numeric columns
print(car_sales.describe())
```

```python
# Access Sale Price column
print(car_sales['Sale Price'])
```

```python
# Access Sale Price and Commission Earned together
print(car_sales[['Sale Price', 'Commission Earned']])
```

```python
# Show the first row
print(car_sales.iloc[0])
```

```python
# Show first 3 rows
print(car_sales.iloc[0:3])
```

```python
# Show cars with Sale Price greater than 40000
high_sales = car_sales[car_sales['Sale Price'] > 40000]
print(high_sales)
```

```python
# Sort by Sale Price from highest to lowest
sorted_sales = car_sales.sort_values(by='Sale Price', ascending=False)
print(sorted_sales)
```

```python
# Sort by Commission Earned from highest to lowest
sorted_commission = car_sales.sort_values(by='Commission Earned', ascending=False)
print(sorted_commission)
```

```python
car_sales.sort_values(by=['Sale Price','Commission Earned'], ascending=[False, False])
```

```python
# Net Price = Sale Price - Commission Earned
car_sales['Net Price'] = car_sales['Sale Price'] - car_sales['Commission Earned']
print(car_sales)
```

```python
# Commission % = (Commission Earned / Sale Price) * 100
car_sales['Commission %'] = (car_sales['Commission Earned'] / car_sales['Sale Price']) * 100

# Show updated DataFrame
print(car_sales)
```

```python
# Group by Salesperson and sum Sale Price
total_sales_person = car_sales.groupby('Salesperson')['Sale Price'].sum()
print(total_sales_person)
```

```python
# Group by Car Make and calculate mean Commission Earned
avg_commission_make = car_sales.groupby('Car Make')['Commission Earned'].mean()
print(avg_commission_make)
```

```python
# Total net price per Car Make
total_net_make = car_sales.groupby('Car Make')['Net Price'].sum()
print(total_net_make)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
print("Seaborn is ready!")

```

```python
# Total sales per salesperson
total_sales_person = car_sales.groupby('Salesperson')['Sale Price'].sum().reset_index()
print(total_sales_person)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

# Bar plot for total sales per salesperson
sns.barplot(data=total_sales_person, x='Salesperson', y='Sale Price', palette="viridis")
plt.title("Total Sales by Salesperson")
plt.show()
```

```python
# Total net price per car make
total_net_make = car_sales.groupby('Car Make')['Net Price'].sum()
print(total_net_make)
```

```python
plt.figure(figsize=(6,6))
plt.pie(total_net_make, labels=total_net_make.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Net Price Contribution by Car Make")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=car_sales, x='Sale Price', y='Commission Earned', hue='Car Make', s=100)
plt.title("Sale Price vs Commission Earned")
plt.xlabel("Sale Price")
plt.ylabel("Commission Earned")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=car_sales, x='Car Make', y='Commission %', palette="coolwarm")
plt.title("Commission % Distribution by Car Make")
plt.xlabel("Car Make")
plt.ylabel("Commission %")
plt.show()
```

```python
# First, make sure Date is in datetime format
car_sales['Date'] = pd.to_datetime(car_sales['Date'])

# Group total sales per date
sales_trend = car_sales.groupby('Date')['Sale Price'].sum().reset_index()
print(sales_trend)
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(data=sales_trend, x='Date', y='Sale Price', marker='o')
plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sale Price")
plt.xticks(rotation=45)
plt.show()
```

```python
sorted_data = car_sales.sort_values(by=['Car Make', 'Sale Price'], ascending=[True, False])
print(sorted_data[['Car Make', 'Car Model', 'Sale Price']])
```

```python
tesla_high_sales = car_sales[(car_sales['Car Make'] == 'Tesla') & (car_sales['Sale Price'] > 40000)]
print(tesla_high_sales[['Car Make', 'Car Model', 'Sale Price', 'Commission Earned']])
```

```python
sales_summary = car_sales.groupby('Salesperson').agg(
    Total_Sales=('Sale Price', 'sum'),
    Avg_Commission=('Commission Earned', 'mean'),
    Max_Sale=('Sale Price', 'max')
).reset_index()

print(sales_summary)
```

```python
car_sales['High_Value'] = car_sales['Sale Price'].apply(lambda x: 'Yes' if x > 40000 else 'No')
print(car_sales[['Car Make', 'Car Model', 'Sale Price', 'High_Value']])
```
